```python
import pandas as pd
requests = pd.read_csv('311requests.csv')
requests.head()
```

```python
income = pd.read_csv('WardIncome.csv')
income
```

```python
parking_requests = requests.loc[requests["SERVICECODEDESCRIPTION"]=='Parking Enforcement']

parking_requests[['X', 'Y', 'SERVICECODEDESCRIPTION', 'WARD']].head()
```

```python
#Group and count each parking request by ward

counted_by_ward = pd.DataFrame(parking_requests[['WARD', 'SERVICECODEDESCRIPTION']]).groupby('WARD').agg('count')
counted_by_ward.reset_index(inplace=True)
counted_by_ward.columns = ['Ward', 'Request Count']
counted_by_ward
```

```python
#Merge the median and mean dataframe with this new dataframe
merged_df = pd.merge(counted_by_ward, income, how='inner', on='Ward')
merged_df
```

This scatterplot shows the relationship between the number of parking enforcement requests and mean household income by ward. This could be due to higher income individuals/families finding other ways of transportation or having personal drivers, etc.

```python
resolved_requests = requests[requests['RESOLUTIONDATE'] != '1970-01-01 00:00:00']

resolved_requests['ADDDATE'] = pd.to_datetime(resolved_requests['ADDDATE'])
resolved_requests['RESOLUTIONDATE'] = pd.to_datetime(resolved_requests['RESOLUTIONDATE'])

resolved_requests['TimeToResolve'] = (resolved_requests['RESOLUTIONDATE'] - resolved_requests['ADDDATE']).dt.total_seconds() / 3600  # in hours

parking_enforcement_requests = resolved_requests[resolved_requests['SERVICECODEDESCRIPTION'] == 'Parking Enforcement']

counted_by_ward = pd.DataFrame(parking_enforcement_requests[['WARD', 'TimeToResolve']])

aggregated_data = counted_by_ward.groupby('WARD')['TimeToResolve'].median().reset_index()
aggregated_data.columns = ['Ward', 'MedianTimeToResolve']
```

```python
income_merged = pd.merge(aggregated_data, income, how='inner', on='Ward')
income_merged
```
