```python
%matplotlib inline

# YOUR CODE HERE
import pandas as pd
import matplotlib.pyplot as plt

orders = pd.read_csv('orders_sample.csv')
order_products = pd.read_csv('order_products_sample.csv')
products = pd.read_csv('products.csv')

product_name = "Bananas"
product_id = products[products['product_name'] == product_name]['product_id'].values[0]

merged_data = pd.merge(orders, order_products, on='order_id')

banana_orders = merged_data[merged_data['product_id'] == product_id]

hourly_banana_orders = banana_orders.groupby('order_hour_of_day')['order_id'].count()

plt.figure(figsize=(10, 6))
plt.plot(hourly_banana_orders.index, hourly_banana_orders.values, marker='o', linestyle='-')
plt.title(f'Orders of {product_name} by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.xticks(range(24))

plt.show()


#raise NotImplementedError()
```

```python
%matplotlib inline

# YOUR CODE HERE
import pandas as pd
import matplotlib.pyplot as plt

orders = pd.read_csv('orders_sample.csv')
order_products = pd.read_csv('order_products_sample.csv')
products = pd.read_csv('products.csv')
departments = pd.read_csv('departments.csv')

merged_data = pd.merge(orders, order_products, on='order_id')
merged_data = pd.merge(merged_data, products, on='product_id')
merged_data = pd.merge(merged_data, departments, on='department_id')

frozen_department = merged_data[merged_data['department'] == 'frozen']

hourly_frozen_orders = frozen_department.groupby('order_hour_of_day')['order_id'].count()

plt.figure(figsize=(10, 6))
hourly_frozen_orders.plot(kind='bar', color='blue')
plt.title('Number of Frozen Food Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.grid(axis='y')

plt.show()




#raise NotImplementedError()
```

```python
%matplotlib inline

# YOUR CODE HERE
import pandas as pd
import matplotlib.pyplot as plt

orders = pd.read_csv('orders_sample.csv')
order_products = pd.read_csv('order_products_sample.csv')
products = pd.read_csv('products.csv')
departments = pd.read_csv('departments.csv')

merged_data = pd.merge(orders, order_products, on='order_id')
merged_data = pd.merge(merged_data, products, on='product_id')
merged_data = pd.merge(merged_data, departments, on='department_id')

deli_department = merged_data[merged_data['department'] == 'deli']

hourly_deli_orders = deli_department.groupby('order_hour_of_day')['order_id'].count()

plt.figure(figsize=(10, 6))
hourly_deli_orders.plot(kind='bar', color='green')
plt.title('Number of Deli Orders by Hour of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.grid(axis='y')

# Show the bar chart
plt.show()
```
